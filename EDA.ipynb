### Importing the libraries and the data

```python
#Import all relevant libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\nikip\Documents\2022\TCE\EDA_main\ToyotaCorolla_eda.csv")
```

```python
df.head()
```

### Line Plot

```python
sns.lineplot(data=df, x="Age", y="Price", hue="FuelType")
```

### Scatter Plot:
- Scatter plots can be used to show a linear relationship between two or three data points using the seaborn library. 
- A Scatter plot of price vs age with default arguments will be like this:

```python
plt.style.use("ggplot")
plt.figure(figsize=(8,6))
sns.regplot(x = df["Age"], y = df["Price"])
plt.show()
```

- Age and Price are 2 numerical variables so, what if you want to add one more categorical variable.
- So there is a function in seaborn libraries called lmplot this helps us to add another categorical variable in the numerical variables Scatter Plot.

```python
sns.lmplot(x='Age', y='Price', data=df,
          fit_reg=False,
          hue='FuelType',
          legend=True,
          palette="Set1",height=6)
```

### Histogram

- A histogram is a great way to visualize the frequency distribution of numerical data. 

```python
plt.figure(figsize=(8,6))
sns.distplot(df['Age'])
plt.show()
```

- If you want to organize all the different intervals or bins, you can use the bins parameter on the distplot function. Let’s use bins = 5 on the distplot function.
- It will organize your bins into five bins or intervals.

```python
plt.figure(figsize=(8,6))
sns.distplot(df['Age'],kde=False,bins=5)
plt.show()
```

```python
plt.figure(figsize=[8,5])
sns.histplot(data=df,x="Price",bins=40).set(title="Distribution of Price",ylabel="Frequency")
plt.show()
```

- Now you can say that from age 65 to 80 we have more than 500 cars.

### Bar Plot:

- Bar plot is for categorical variables. Bar plot is the commonly used plot because of its simplicity and it’s easy to understand data through them. 
- You can plot a barplot in seaborn using the countplot library.
- It’s really simple. Let’s plot a barplot of FuelType.

```python
plt.figure(figsize=(8,6))
sns.countplot(x="FuelType", data=df)
plt.show()
```

### Grouped Bar Plot:

- We can plot a barplot between two variables. 
- That’s called grouped barplot. 
- Let’s plot a barplot of FuelType distributed by different values of the Automatic column.

```python
plt.figure(figsize=(8,6))
sns.countplot(x="FuelType", data=df,
             hue="Automatic")
plt.show()
```

### Box and Whiskers Plot:

- Box and whiskers plots are used for analyzing the detailed distribution of a dataset. 
- Let’s plot Box and whiskers plot of the Price column of the dataset to visually interpret the “five-number summary”. 
- Five Number Summary includes minimum, maximum, and the three quartiles(1st quartile, median and third quartile).

```python
plt.figure(figsize=(8,6))
sns.boxplot(y=df["Price"])
plt.show()
```

- Anything above the whiskers is called Outliers. Outlier or extreme values lies above 1.5 times the median values.

### Box and Whiskers Plot(Numerical vs Categorical Variable): 

```python
plt.figure(figsize=(8,6))
sns.boxplot(x=df["FuelType"], 
           y=df["Price"],
           )
plt.show() 
```

### Correlation Matrix

- Use a correlation matrix to get more specific information about the relationship between these two variables. A correlation matrix is useful for identifying the relationship between several variables.

```python
#Create a correlation matrix to show relationship between select variables
corr_matrix = df[["Age","Price","KM", "FuelType","HP","Automatic" , "CC", "Doors", "Weight" ]].corr()
#corr_matrix

#Create a heatmap to visualize correlation
plt.figure(figsize=[12,9])
sns.heatmap(corr_matrix,annot=True,cmap='Reds')
plt.title("Correlation between Selected Variables")
plt.show()
```

### Pairwise Plots:

- A pair plot is used to plot pairwise relationships between columns in a dataset. 
- Create scatterplots for joint relationships and histograms for univariate distribution or relationships.  
- It will show the relationship between all the different variables in a particular dataset. 
- In a pair plot, we can pass hue as a parameter. hue is the parameter on which we want to calculate the pairwise plot. Let’s plot a pair plot by passing FuelType as a hue. 
- Here hue = FuelType because we want to do this pairwise plot by using FuelType on all the rest of the columns.

```python
sns.pairplot(df, kind="scatter", hue="FuelType")
plt.show()
```
